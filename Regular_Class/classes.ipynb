```python
# restuarant information
# customers coming in placing orders of menu items
# collecting orders
# billing
# print bill
# update order
# search for an order

import random

class restuarant:

    def __init__(self):

        self.menu = {
            "rice" : 30,
            "idli" : 10
        }

        self.order = {}

    def print_menu(self):
        print('The current options in the menu are: ')
        print(self.menu)
        print('*' * 30)

    def generate_orderid(self):
        order_id = ""
        for i in range(1,6,1):
            order_id += str(random.randint(i,10))
        return order_id

    def collectorder(self):
        self.print_menu()
        phone = input('Enter customer number: ')
        order = { }
        
        while True:
            item = input('Enter the item name you want to order: ')
            qty = int(input('Enter the quantity of item: '))
            order[item] = qty

            choice = input("Do you want to complete the order? Y or N")
            if choice == 'Y':
                break
            
        order_id = self.generate_orderid()
        self.order[order_id] = {

            "Phone" : phone,
            "Order" : order

        }

        print(self.order)

    def billing(self):
        total = 0
        totalEarn = 0
        for orderid in self.order:
            print(self.order[orderid]['Order'])
            for key in self.order[orderid]['Order']:
                print(key)
                print(self.order[orderid]['Order'][key]) #qty
                print(self.menu[key]) #unit price
                price = self.order[orderid]['Order'][key] * self.menu[key]
                total = total+price
                totalEarn = totalEarn+total
                print("price = ",price)
            print("total = ",total)
        print("Earning of today is: ",totalEarn,"Rupees")
        # print menu
        # ask user to enter the value
        # collect the value and update the order
        # print the order

cbtv = restuarant()
cbtv.collectorder()
cbtv.billing()
```
