write a python program with the following specification there will be a function named create dataset which will take a dictionary,filename,number of rows as arguements. the dictionary is having the following structure{
    "colunmname":[domain of values]
}
your function will select random values for each of the domains  and create a row for you .
the number of rows that has to be crated is a parameter that has been taken. save the file with the number of rows.
load this dataset using pandas and display basic information

```python
import random

# Creating dictionary

dict = {
    'Name' : ['Repiece','Devu','Kundimwon','Shaddimwon','Kusu','Jof'] ,
    'Age' : ['23','22','21','20','19','25'] ,
    'Year' : ['2000','2002','2003','2016','2017','2019'] ,
    'Year of Study' : ['2','4','5','4','1','6'] ,
    'Income' : ['0'] ,
    'Place' : ['Kerala']
}

# Creating a function CreateDataset

def createDataset(filename,n):

    file = open(filename,'a+')
    file.write('Name' '\t' + 'Age' '\t' + 'Year' '\t' + 'Year of study' '\t' + 'Income' '\t' + 'Place' '\n')
    file.close()

    file = open(filename,'a+' )
    for i in range(n+1):
        file.write(random.choice(dict['Name']) + '\t' + random.choice(dict['Age']) + '\t' + random.choice(dict['Year']) + '\t' + random.choice(dict['Year of Study']) + '\t' + random.choice(dict['Income']) + '\t' + random.choice(dict['Place']) + '\n')
    file.close()

createDataset('Unda.csv' , 20)
```
